```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('water_potability.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Check for missing values...
df.isnull().sum()
```

```python
# Fill missing values with median values...
for col in df.columns:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
```

```python
# Recheck for missing values...
df.isnull().sum()
```

```python
# Check for duplicate values...
df.duplicated().sum()
```

```python
#Saving the cleaned data...
df.to_csv('water_potability_cleaned.csv', index=False)
```

```python
# Data Analysis
# (i) Distribution between Unsafe / Safe data samples...
sns.countplot(x='Potability', data=df)
plt.title('Distribution of Potability (0=Not Safe, 1=Safe)')
plt.show()
```

```python
# (ii) Histograms for each feature...
df.hist(figsize=(12, 10), bins=30)
plt.tight_layout()
plt.show()
```

```python
# (iii) Boxplots...
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop('Potability', axis=1))
plt.xticks(rotation=90)
plt.show()
```

```python
# (iv) Heatmap...
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between features')
plt.show()
```
